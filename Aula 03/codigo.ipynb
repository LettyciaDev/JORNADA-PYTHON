```python
# Passo a passo
# Passo 0: Entender o desafio da empresa e da sua rea
# Passo 1: Importar a base de dados

import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
# verificar se a base de dados tem problemas
display(tabela.info())

# float
# int
# object --> precisa ser modificada para nmero, pois a IA s consegue aprender com informaes nmericas ('Preprocessing')
```

```python
# Passo 2: Preparar a base de dados para a inteligencia artificial

# colunas que so texto:
#NICA COLUNA QUE PODE ESTAR EM FORMATO DE TEXTO  A SCORE_CREDITO, POIS IR IMPRIMIR O RESULTADO DA APROVAO DE CRDITO

    
# Label Encoder: processo que codifica com um rotulo a cada object, atribuindo um nmero
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()

# profissao
    # cientista - 1
    # ator - 2
    # mecanicico - 3
    # engenherio - 4

tabela["profissao"] = codificador.fit_transform(tabela["profissao"])

# mix_credito
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])

# comportamento_pagamento
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])

display(tabela.info())

```

```python
# Antes de apresentar os dados para a IA,  preciso dividir as colunas em duas partes:
    #  A parte da coluna que vai ser prevista (recebe o nome de y) 
     # A parte das colunas que ajudaro a fazer a previso (recebe o nome de x)

x = tabela.drop(columns=["score_credito", "id_cliente"])
y = tabela["score_credito"]

# Em seguida, dividir os dados em dados para treino(x e y) e dados para teste(x e y),
# a fim de verificar se a IA realmente aprendeu a fazer previses

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# Passo 3: Criar um modelo de IA --> Nota de credito: ruim, mdia e boa

# rvore de deciso -> RandomForest : treina baseado em perguntas
# Neirest Neighbors -> KNN : treina baseado proximidades de cada elemento

# 3 etapas para criar uma IA:

# Importar a IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar a IA
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#treinar a IA
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Passo 4: Escolher o melhor modelo 

#pedir para os modelos previrem os creditos(y_teste) atravs dos x_teste e comparar os resultados
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score

#percentual de acertos
display(accuracy_score(y_teste, previsao_arvoredecisao))
display(accuracy_score(y_teste, previsao_knn))
```

```python
# Passo 5: Usar o nosso modelo para fazer novas previses

# Importar novos clientes
tabela_novos_clientes = pd.read_csv("novos_clientes.csv")
display(tabela_novos_clientes)

# Visualizar tabela

# fazer pre tratamento

tabela_novos_clientes["profissao"] = codificador.fit_transform(tabela_novos_clientes["profissao"])

tabela_novos_clientes["mix_credito"] = codificador.fit_transform(tabela_novos_clientes["mix_credito"])

tabela_novos_clientes["comportamento_pagamento"] = codificador.fit_transform(tabela_novos_clientes["comportamento_pagamento"])

#fazer previses
previsoes = modelo_arvoredecisao.predict(tabela_novos_clientes)
display(previsoes)
```
